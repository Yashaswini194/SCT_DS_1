```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv(r'C:\Users\YASHASWINI\OneDrive\Desktop\books\books.csv')

# Clean rating column
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')
df = df.dropna(subset=['Rating'])

# Create bins for rating categories
bins = [0, 2, 3, 4, 5]
labels = ['Low (0–2)', 'Average (2–3)', 'Good (3–4)', 'Excellent (4–5)']

df['Rating_Group'] = pd.cut(df['Rating'], bins=bins, labels=labels, right=True)

# Count values
counts = df['Rating_Group'].value_counts().sort_index()

# Plot
plt.figure(figsize=(10,6))
bars = plt.bar(counts.index, counts.values, color=['#ff9999','#ffcc66','#66b3ff','#99ff99'])

# Add values on top of bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height,
             f'{int(height)}',
             ha='center', va='bottom', fontsize=10)

# Labels and title
plt.title("Distribution of Book Ratings", fontsize=16, fontweight='bold')
plt.xlabel("Rating Category")
plt.ylabel("Number of Books")
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```
